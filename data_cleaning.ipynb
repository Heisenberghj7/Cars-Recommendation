## 2- Some Basic Cleaning

```python
# We should know what the values look like in each column in order to know what type of cleaning they need
```

```python
def remove_newlines(input_string):
    output_string = input_string.replace("\n", "")
    return output_string
```

```python
New_Cars["Brand & Model"]=New_Cars["Brand & Model"].apply(remove_newlines)
```

```python
def Price_cleaning(input_string):
    to_delete=["DH","promo","*","Prix","public","\n"]
    strings=input_string.split(" ")
    for string in strings:
        if string in to_delete:
            input_string=input_string.replace(string,"")
    return input_string.strip()
```

```python
def correct_number(input_num):
    if input_num != '':
        input_num=input_num.replace('.', '')
    else: 
        input_num = 0
    return input_num
```

```python
New_Cars["Original Price"]= New_Cars["Original Price"].apply(Price_cleaning)
New_Cars["Promo"]= New_Cars["Promo"].apply(Price_cleaning)
```

```python
New_Cars["Original Price"]= New_Cars["Original Price"].apply(correct_number)
New_Cars["Promo"]= New_Cars["Promo"].apply(correct_number)
```

```python
New_Cars["Original Price"]= New_Cars["Original Price"].astype(float)
New_Cars["Promo"]= New_Cars["Promo"].astype(float)
```

```python
New_Cars["Energie"].value_counts()
```

```python
def cylindre(input_text):
    input_text=input_text.replace("cm³","").strip()
    return input_text
```

```python
New_Cars["Cylindrée"]=New_Cars["Cylindrée"].astype(str)
New_Cars["Cylindrée"]=New_Cars["Cylindrée"].apply(cylindre)
```

```python
New_Cars["Puissance fiscale"]=New_Cars["Puissance fiscale"].astype(str)
New_Cars["Puissance fiscale"]=New_Cars["Puissance fiscale"].apply(lambda x: x.replace("cv",""))
```

```python
New_Cars["Puissance maxi."]=New_Cars["Puissance maxi."].astype(str)
New_Cars["Puissance maxi."]=New_Cars["Puissance maxi."].apply(lambda x: x.replace("ch",""))


New_Cars["Couple maxi."]=New_Cars["Couple maxi."].astype(str)
New_Cars["Couple maxi."]=New_Cars["Couple maxi."].apply(lambda x: x.replace("Nm",""))
```

```python
New_Cars=New_Cars.rename(columns = {"Cylindrée":"Cylindrée (cm³)","Puissance fiscale":"Puissance fiscale (cv)",
                                    "Puissance maxi.":"Puissance maxi (ch)","Couple maxi.":"Couple maxi (Nm)"})
```
