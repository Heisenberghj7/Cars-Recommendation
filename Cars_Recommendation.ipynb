### Content Based Recommendation System

```python
New_Cars.tail()
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Concatenate relevant columns into a single feature
New_Cars['features'] = New_Cars.apply(lambda row: ' '.join(row.values.astype(str)), axis=1)

# Create a TF-IDF vectorizer
tfv = TfidfVectorizer(min_df=10,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')

# Compute TF-IDF matrix
tfidf_matrix = tfv.fit_transform(New_Cars['features'])

# Compute the cosine similarity matrix
cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)

# Function to get recommendations based on similarity scores
def get_recommendations(name, cosine_similarities, data, top_n=15):
    # Get the index of the car with the given name
    idx = data[data['name'] == name].index[0]
    
    # Get the similarity scores for all cars
    similarity_scores = list(enumerate(cosine_similarities[idx]))
    
    # Sort the cars based on similarity scores
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    
    # Get the top N most similar cars
    top_similar_cars = [data.iloc[i[0]]['name'] for i in similarity_scores[1:top_n+1]]
    
    return top_similar_cars

# Example usage
car_name = 'VOLVO XC40 2.0 B3 163 MHEV Plus Dark'
recommendations = get_recommendations(car_name, cosine_similarities, New_Cars)
print(f"Recommendations for {car_name}:")
for car in recommendations:
    print(car)
```
